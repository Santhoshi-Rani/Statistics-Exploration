```python
# for outlier treatment lets import another dataset
out_df = pd.read_csv('winequality.csv',sep=';')
out_df
```

```python
out_df.boxplot()
```

```python
out_df.boxplot('total sulfur dioxide')  # Before Oulier Treatment
```

```python
Q1 = out_df['total sulfur dioxide'].quantile(0.25)
Q3 = out_df['total sulfur dioxide'].quantile(0.75)

IQR = Q3 - Q1

LL = Q1 - (1.5 * IQR)
UL = Q3 + (1.5 * IQR)

print(f"Q1: {Q1} | Q3: {Q3} | IQR: {IQR} | LL: {LL} | UL: {UL}")
```

```python
ul_outliers = out_df[out_df['total sulfur dioxide'] > UL].shape[0]
ll_outliers = out_df[out_df['total sulfur dioxide'] < LL].shape[0]

total_outliers = ul_outliers + ll_outliers

perc_outliers = total_outliers * 100 / out_df.shape[0]

print(f"UL Count: {ul_outliers} | LL Count: {ll_outliers} | Total: {total_outliers} | Total%: {perc_outliers}")
```

```python
out_df.loc[out_df['total sulfur dioxide'] > UL] = UL
out_df.loc[out_df['total sulfur dioxide'] < LL] = LL
```

```python
out_df.boxplot('total sulfur dioxide')  # After Outlier Treatment
```
